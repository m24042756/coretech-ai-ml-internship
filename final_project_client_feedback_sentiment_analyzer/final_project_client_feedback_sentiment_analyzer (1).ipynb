```python
import pandas as pd

df = pd.read_csv("coretech_feedback.csv")

df.head()
```

```python
print(df.info())
```

```python
print(df.isnull().sum())
```

```python
import matplotlib.pyplot as plt

df["Sentiment"].value_counts().plot(kind="bar")

plt.title("Sentiment Distribution")

plt.show()
```

```python
df["Rating"].hist()

plt.title("Rating Distribution")

plt.show()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer()

X = tfidf.fit_transform(df["Feedback_Text"])

y = df["Sentiment"]

print(X.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)

print("Train/Test Split Completed")
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score

nb = MultinomialNB()

nb.fit(X_train, y_train)

nb_pred = nb.predict(X_test)

nb_accuracy = accuracy_score(y_test, nb_pred)

print("Naive Bayes Accuracy:", nb_accuracy)
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()

lr.fit(X_train, y_train)

lr_pred = lr.predict(X_test)

lr_accuracy = accuracy_score(y_test, lr_pred)

print("Logistic Regression Accuracy:", lr_accuracy)
```

```python
comparison = pd.DataFrame({
    "Model": ["Naive Bayes", "Logistic Regression"],
    "Accuracy": [nb_accuracy, lr_accuracy]
})

print(comparison)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, lr_pred))
```

```python
import matplotlib.pyplot as plt

comparison.plot(
    x="Model",
    y="Accuracy",
    kind="bar"
)

plt.title("Model Accuracy Comparison")
plt.ylabel("Accuracy")

plt.show()
```

## Service-wise Feedback Count

This chart shows the number of feedback entries received for each service category.
It helps identify which services generate the most client feedback.

```python
# Service-wise Feedback Count

df["Service"].value_counts().plot(kind="bar")

plt.title("Service-wise Feedback Count")

plt.xlabel("Service")

plt.ylabel("Count")

plt.show()
```

## Sentiment by Service

This chart compares Positive, Neutral, and Negative feedback across different services.
It helps identify which services have higher customer satisfaction and which require improvement.

```python
# Sentiment by Service

pd.crosstab(
    df["Service"],
    df["Sentiment"]
).plot(kind="bar")

plt.title("Sentiment by Service")

plt.xlabel("Service")

plt.ylabel("Count")

plt.show()
```

## Average Rating by Service

This chart shows the average rating received by each service.
Higher average ratings indicate better customer satisfaction and service quality.

```python
# Average Rating by Service

df.groupby("Service")["Rating"].mean().plot(kind="bar")

plt.title("Average Rating by Service")

plt.ylabel("Average Rating")

plt.show()
```
