```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
```

## Creating Sample Client Dataset

```python
data = {
    "Client_ID":[1,2,3,4,4,5],
    "Name":["Ali","Sara","Ahmed","Ayesha","Ayesha",None],
    "City":["Karachi","Lahore","Karachi","Islamabad","Islamabad","Lahore"],
    "Age":[22,25,None,28,28,30],
    "Salary":[50000,60000,55000,70000,70000,None],
    "Subscription":["Basic","Premium","Basic","Premium","Premium","Basic"]
}

df = pd.DataFrame(data)

print("Original Dataset")
print(df)
```

## Checking Missing Values

```python
print(df.isnull().sum())
```

## Handling Missing Values

```python
df["Age"] = df["Age"].fillna(df["Age"].mean())
df["Salary"] = df["Salary"].fillna(df["Salary"].mean())
df["Name"] = df["Name"].fillna("Unknown")

print(df)
```

## Removing Duplicate Records

```python
print("Rows Before Removing Duplicates:", len(df))

df = df.drop_duplicates()

print("Rows After Removing Duplicates:", len(df))
print(df)
```

## Label Encoding

```python
le = LabelEncoder()

df["Subscription"] = le.fit_transform(df["Subscription"])

print(df)
```

## One Hot Encoding

```python
df = pd.get_dummies(df, columns=["City"])

print(df)
```

## Before Scaling

```python
print(df[["Age","Salary"]])
```

## Feature Scaling Using MinMaxScaler

```python
scaler = MinMaxScaler()

df[["Age","Salary"]] = scaler.fit_transform(df[["Age","Salary"]])

print(df[["Age","Salary"]])
```

## Feature Selection

```python
selected_features = df[["Age","Salary","Subscription"]]

print("Selected Features")
print(selected_features)
```

## Final Cleaned Dataset

```python
print(df)
```

## Saving Cleaned Dataset

```python
df.to_csv("cleaned_clients_dataset.csv", index=False)

print("Cleaned dataset saved successfully.")
```
